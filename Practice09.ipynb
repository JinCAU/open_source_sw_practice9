```python
import sklearn.datasets
import sklearn.linear_model
import sklearn.model_selection
import sklearn.metrics
import numpy as np

import matplotlib.pyplot as plt 
%matplotlib inline
```

# Generate Data Points (Do not modify the following block)

```python
xmin, xmax = -5, 5
n_samples = 100
np.random.seed(0)
X = np.random.normal(size=n_samples)
y = (X > 0).astype(float)
X[X > 0] *= 4
X += 0.7 * np.random.normal(size=n_samples)
X = X[:, np.newaxis]
```

### 1. Plot data points

```python
plt.scatter(X, y, label="Training Data")
plt.show()
```

### 2. Create logistic regression object using scikit learn with default hyper parameters
hint: sklearn.linear_model.LogisticRegression()

```python
logistic_regression = sklearn.linear_model.LogisticRegression()
```

### 3. Train a logistic regression for training dataset
hint: .fit(,) 

```python
logistic_regression.fit(X, y)
```

### 4. Compute the predicted probability of the trained logistic regression
hint: .predict_proba()
- predict labels for test dataset (X_test)
- store the predicted probability of X_test to the variable 'y_pred'
- 'y_pred' was initialized as zero vectors and fill 'y_pred' with predicted labels
- do not change the variable name, 'y_pred'

```python
# TO DO
X_test = np.linspace([-5], [10], 300)
y_pred=logistic_regression.predict_proba(X_test)
print(y_pred.shape)
```

### 5. Plot predicted probabilities with data points

```python
plt.figure(1, figsize=(8, 6))
plt.scatter(X, y, label="Training Data")
plt.plot(X_test, y_pred[:, 0], label="Predicted Probabilities(Y=0)")
plt.plot(X_test, y_pred[:, 1], label="Predicted Probabilities(Y=1)")
plt.ylabel("y")
plt.xlabel("X")
plt.xticks(range(-5, 10))
plt.yticks([0, 0.5, 1])
plt.ylim(-0.25, 1.25)
plt.xlim(-4, 10)
plt.legend()
plt.tight_layout()
plt.show()
```

### 8. Load iris dataset from scikit learn and train the logistic regression, and change at least one hyperparameter

```python
iris=sklearn.datasets.load_iris()
X=iris.data
y=iris.target
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)

logistic_regression3=sklearn.linear_model.LogisticRegression(max_iter=10000)
logistic_regression3.fit(X_train, y_train)
y_pred3=logistic_regression3.predict(X_test)
print('Accuracy: %.2f' % sklearn.metrics.accuracy_score(y_test, y_pred3))
```

### 9. Load breast cancer dataset from scikit learn, train the logistic regression, and change at least one hyperparameter
hint: sklearn.datasets.load_breast_cancer()

```python
boston=sklearn.datasets.load_breast_cancer()
X=boston.data
y=boston.target
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)

logistic_regression4=sklearn.linear_model.LogisticRegression(tol=0.1)
logistic_regression4.fit(X_train, y_train)
y_pred4=logistic_regression4.predict(X_test)
print('Accuracy: %.2f' % sklearn.metrics.accuracy_score(y_test, y_pred4))
```

### 10. Load wine dataset from scikit learn, train the logistic regression, and change at least one hyperparameter
hint: sklearn.datasets.load_wine()

```python
wine=sklearn.datasets.load_wine()
X=wine.data
y=wine.target
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)

logistic_regression5=sklearn.linear_model.LogisticRegression(intercept_scaling=0.1)
logistic_regression5.fit(X_train, y_train)
y_pred5=logistic_regression5.predict(X_test)
print('Accuracy: %.2f' % sklearn.metrics.accuracy_score(y_test, y_pred5))
```
